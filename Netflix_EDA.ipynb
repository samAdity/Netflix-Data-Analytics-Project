```python
df=df.dropna()
df.shape
```

```python
df.head(5)
```

```python
df["date_added"] = pd.to_datetime(df['date_added'].str.strip(), format='%B %d, %Y', errors='coerce')
df['day_added'] = df['date_added'].dt.day
df['year_added'] = df['date_added'].dt.year
df['month_added']=df['date_added'].dt.month
df['year_added'].astype(int);
df['day_added'].astype(int);
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))  # Set figure size

sns.countplot(x="type", data=netflix, palette="Set3")  # Use x-axis & custom colors

plt.title("Type")  # Set title
plt.show()
```

```python
plt.figure(figsize=(13, 13))  # Set figure size

# Plot countplot with vertical bars and a color palette
ax = sns.countplot(x="rating", data=netflix, palette="Set2")

# Rotate x-axis labels for better readability
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")

plt.title("Rating")  # Set title
plt.show()
```

```python
plt.figure(figsize=(10, 8))  # Set figure size

# Use 'x' for a vertical bar chart
ax = sns.countplot(x='rating', hue='type', data=netflix, palette="Set2")

# Rotate x-axis labels for better readability
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")

plt.title("Relation between Type and Rating")  # Set title
plt.show()
```

```python
labels = ['Movie', 'TV show']
size = netflix['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size,labels=labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type', fontsize = 25)
plt.legend()
plt.show()
```

```python
netflix['rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.show()
```
